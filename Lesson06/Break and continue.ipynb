# Solutions for break and continue exercises

1. Write a string of odd numbers in range of 20, by using `continue`.

```python
for n in range(20):
    #if the remainder of n / 2 is 1, skip the rest to the loop.
    if n % 2 == 0:
        continue
    print(n, end=' ')
```
