**Question 1**

Create a cosine values of data that range from 0 to 10 by using numpy.
Use `np.linspace()` to get the range value from 0 to 10. Use `np.cos()` to get the cosine values for the value above. Then use `plt.plot()` to plot the line graph for the cosine values above.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(1)
x = np.linspace(0,10)
y = np.cos(x)
plt.plot(x,y)
```
