**Question 1**

Create a data frame (table of data with row and column) by using `pd.DataFrame()`.
The data should contain:

`"label": ["A", "B", "A", "B", "A", "B"]`

`"value": [1, 2, 3, 4, 5, 6]`

```python
import pandas as pd
df = pd.DataFrame({"label":["A", "B", "A", "B", "A", "B"],
                  "value": [1, 2, 3, 4, 5, 6]})
df
```

**Question 2**

**Question 3**

```python
df["label"].str.lower()
```

**Question 4**

Calculate the sum of values in column `value`.

```python
df["value"].sum()
```

**Question 5**

Calculate the sum of values according to their labels. by using `df.groupby().sum()`

```python
df.groupby("label").sum()
```
